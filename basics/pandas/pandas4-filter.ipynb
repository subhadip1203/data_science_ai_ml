```python
import pandas as pd
pokemon = pd.read_csv("pokemon_data.csv")
pokemon
```

```python
pokemon.loc[ pokemon["Type 1"]=="Grass" ]
```

```python
# when we do filtering
# old index is still there : the first column
newPokemon = pokemon.loc[ (pokemon["Type 1"]=="Grass") & (pokemon["Type 2"]=="Poison")]
newPokemon
```

```python
# to reset index
newPokemon = newPokemon.reset_index(drop=True)
newPokemon
```

```python
megaPokemon = pokemon.loc[pokemon["Name"].str.contains("Mega")  ]
megaPokemon
```

```python
# NOT sign is not ! , but ~
megaPokemon = pokemon.loc[~pokemon["Name"].str.contains("Mega")  ]
megaPokemon
```

```python
# Regex expression
import re
megaPokemon = pokemon.loc[~pokemon["Type 1"].str.contains("Fire|Grass", regex=True)  ]
megaPokemon
```
